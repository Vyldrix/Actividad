```python
%%writefile primo_linux.asm
; primo_linux.asm - Determina si 17 es primo en Linux x86_64
; Compilar: nasm -f elf64 primo_linux.asm -o primo_linux.o
; Enlazar: ld primo_linux.o -o primo_linux

section .data
    num     dq 18                      ; Número a verificar
    msgPrimo    db "El numero es primo", 10
    msgPrimoLen equ $-msgPrimo
    msgNoPrimo  db "El numero NO es primo", 10
    msgNoPrimoLen equ $-msgNoPrimo

section .text
global _start

_start:
    mov rax, [num]    ; rax = número a verificar
    mov rcx, 2        ; divisor inicial

check_loop:
    cmp rcx, rax
    jge es_primo       ; si divisor >= número → primo

    xor rdx, rdx       ; limpiar resto
    mov rbx, rcx
    mov rax, [num]
    div rbx            ; rax / rbx → cociente en rax, resto en rdx

    cmp rdx, 0
    je no_primo        ; resto 0 → no primo

    inc rcx
    jmp check_loop

es_primo:
    mov rax, 1         ; syscall: sys_write
    mov rdi, 1         ; stdout
    lea rsi, [rel msgPrimo]
    mov rdx, msgPrimoLen
    syscall
    jmp salir

no_primo:
    mov rax, 1
    mov rdi, 1
    lea rsi, [rel msgNoPrimo]
    mov rdx, msgNoPrimoLen
    syscall

salir:
    mov rax, 60        ; syscall: sys_exit
    xor rdi, rdi
    syscall
```

```python
!nasm -f elf64 primo_linux.asm -o primo_linux.o
!ld primo_linux.o -o primo_linux
```
