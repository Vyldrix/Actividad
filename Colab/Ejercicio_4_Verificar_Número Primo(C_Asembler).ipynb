```python
%%writefile primo.c
#include <stdio.h>

int main() {
    long num = 17;      // NÃºmero a verificar
    long i = 2;         // Divisor inicial

    if (num < 2) {
        printf("El numero NO es primo\n");
        return 0;
    }

    while (i < num) {
        if (num % i == 0) {
            printf("El numero NO es primo\n");
            return 0;
        }
        i++;
    }

    printf("El numero es primo\n");
    return 0;
}
```
